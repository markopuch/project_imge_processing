```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
import glob
```

```python
cv_img = []
samp=8


for img in glob.glob("*.jpg"):
    n= cv2.imread(img)
    n=cv2.cvtColor(n, cv2.COLOR_BGR2RGB)
    n=cv2.resize(n,(1200,700))
    cv_img.append(n)


plt.figure(figsize=(40,40))
for i in range(samp):
    plt.subplot(4,4,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(cv_img[i],cmap='gray')
plt.show()
```

```python

mask=[]
kernel=15
l_w=np.array([0,0,110])
u_w=np.array([0,0,255])

for i in range(samp):
    n=cv_img[i]
    #proces
    n=cv2.cvtColor(n,cv2.COLOR_RGB2GRAY)
    n=cv2.GaussianBlur(n,(kernel,kernel),3)
    n=cv2.cvtColor(n,cv2.COLOR_GRAY2RGB)
    n=cv2.cvtColor(n,cv2.COLOR_RGB2HSV)
    n=cv2.inRange(n,l_w,u_w)

    mask.append(n)


    
plt.figure(figsize=(40,40))
for i in range(samp):
    plt.subplot(4,4,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(mask[i],cmap='gray')
plt.show()
```
