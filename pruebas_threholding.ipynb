```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
import glob
from skimage import exposure
```

```python
img=cv2.imread('samples/test_image_1.jpg',1) #8

print(img.shape)
plt.imshow(img)
plt.show()
```

```python
img=cv2.resize(img,(1200,700))
print (img.shape)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_process=np.copy(img)
plt.imshow(img_process)
plt.show()
```

```python
gray=cv2.cvtColor(img_process,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
kernel=5
gauss_filter = cv2.GaussianBlur(gray,(kernel,kernel),3)
plt.imshow(gauss_filter,cmap='gray')
plt.show()
```

```python
plt.hist(gray.ravel(),256,[0,256]); plt.show()
```

```python
limite=120
ret,bin=cv2.threshold(gauss_filter,limite,255,cv2.THRESH_BINARY )
# cv2.imwrite('filename.jpg', th2)
plt.imshow(bin,cmap='gray')
plt.show()
```

```python
ng=7 #vecindad
c=2 #constante
th3=cv2.adaptiveThreshold(gauss_filter,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,ng,c)
            
plt.imshow(th3,cmap='gray')
plt.show()
th4=cv2.adaptiveThreshold(gauss_filter,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY_INV,ng,c)
            
plt.imshow(th4,cmap='gray')
plt.show()
```

```python
mask=cv2.bitwise_and(bin, th3)
plt.imshow(mask,cmap='gray')
plt.show()
```
